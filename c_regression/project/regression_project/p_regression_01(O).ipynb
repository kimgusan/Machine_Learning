## Cycle01

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

features, targets = pre_l_df.iloc[:, :-1], pre_l_df.iloc[:, -1]

X_train, X_test, y_train, y_test = \
train_test_split(features, targets, test_size=0.2, random_state=124)

l_r = LinearRegression()
l_r.fit(X_train, y_train)
```

```python
import numpy as np
from sklearn.metrics import mean_squared_log_error, mean_squared_error, r2_score

def get_evaluation(y_test, prediction):
    MSE = mean_squared_error(y_test, prediction)
    RMSE = np.sqrt(MSE)
    MSLE = mean_squared_log_error(y_test, prediction)
    RMSLE = np.sqrt(mean_squared_log_error(y_test, prediction))
    R2 = r2_score(y_test, prediction)
    print('MSE: {:.4f}, RMSE: {:.4f}, MSLE: {:.4f}, RMSLE: {:.4f}, R2: {:.4f}'\
          .format(MSE, RMSE, MSLE, RMSLE, R2))
```

```python
prediction = l_r.predict(X_test)
get_evaluation(y_test, prediction)
```

## Cycle03 Í≥ºÏ†ÅÌï©ÏùÑ ÌåêÎã®ÌïòÍ∏∞ ÏúÑÌï¥ ÍµêÏ∞®Í≤ÄÏ¶ù ÏßÑÌñâ

```python
# ÌååÏù¥ÌîÑÎùºÏù∏ Íµ¨Ï∂ï ÌõÑ Ï∞®Ïõê Ï∂ïÏÜå ÌõÑ ÏÑ†Ìòï ÌöåÍ∑Ä Î∂ÑÏÑù
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split, GridSearchCV, KFold
from sklearn.pipeline import Pipeline


features, targets = pre_l_df.iloc[:,:-1], pre_l_df.iloc[:,-1]

X_train, X_test, y_train, y_test = \
train_test_split(features,targets, test_size=0.2, random_state=321)

kfold = KFold(n_splits=15, random_state=321, shuffle=True)

pipe = Pipeline(
    [
        ('pca', PCA(n_components=8)),  # PCAÎ°ú Ï∞®Ïõê Ï∂ïÏÜå
        ('l_r', LinearRegression())    # ÏÑ†Ìòï ÌöåÍ∑Ä Î™®Îç∏
    ]
)

param_grid = {
    'pca__n_components': [8]  # PCA Ïª¥Ìè¨ÎÑåÌä∏ Ïàò Ï°∞Ï†ï
}

grid_l = GridSearchCV(pipe, param_grid=param_grid, cv=kfold, scoring='neg_mean_squared_error')
grid_l.fit(X_train, y_train)

# ÏµúÏ†ÅÏùò ÌååÎùºÎØ∏ÌÑ∞ÏôÄ ÏÑ±Îä• Ï∂úÎ†•
print("Best parameters:", grid_l.best_params_)
print("Best cross-validation score: {:.3f}".format(-grid_l.best_score_))
```

```python
prediction = grid_l.predict(X_test)
get_evaluation(y_test, prediction)
```

## Cycle04
- ÌååÏù¥ÌÜ†ÏπòÎ•º ÏÇ¨Ïö©ÌïòÏó¨ loss Í∞íÏùÑ ÌôïÏù∏ÌïòÏó¨ Í∑∏ÎûòÌîÑÎ•º ÌôïÏù∏ÌïòÏó¨ Ïã§Ï†ú train Ïùò Í∞íÍ≥º test Ïùò Í∞íÏù¥ Ïò§Ï∞®Í∞Ä Ïª§ÏßÄÎäî ÏßÄÏ†êÏùÑ Ï∞æÏïÑÏÑú, Í∑úÏ†úÎ•º ÌôïÏù∏ÌïúÎã§.

```python
import torch
from torch.optim import SGD
from torch.nn.functional import mse_loss
from torch.nn import Linear

torch.manual_seed(321)

features, targets = pre_l_df.iloc[:,:-1], pre_l_df.iloc[:,-1]

X_train, X_test, y_train, y_test = \
train_test_split(features,targets, test_size=0.2, random_state=321)

real_X_train, val_X_test, real_y_train, val_y_test = \
train_test_split(X_train, y_train, test_size=0.5, random_state=321)

real_X_train = torch.FloatTensor(X_train.values)
real_y_train = torch.FloatTensor(y_train.values).view(-1, 1)

val_X_test = torch.FloatTensor(X_test.values)
val_y_test = torch.FloatTensor(y_test.values).view(-1, 1)
l_r = Linear(real_X_train.shape[1], 1)  # ÏûÖÎ†• Ï∞®Ïõê ÎèôÏ†Å Ìï†Îãπ
optimizer = SGD(l_r.parameters(), lr=0.001)  # Î≥¥Îã§ Ïã§Ïö©Ï†ÅÏù∏ ÌïôÏäµÎ•†

epochs = 1000000  # Ï†ÅÏ†àÌïú ÏóêÌè¨ÌÅ¨ Ïàò
real_train_loss_history = []
val_test_loss_history = []

for epoch in range(1, epochs + 1):
    l_r.train()
    H = l_r(real_X_train)
    train_loss = mse_loss(real_y_train, H)

    optimizer.zero_grad()
    train_loss.backward()
    optimizer.step()
    
    l_r.eval()
    with torch.no_grad():
        H_test = l_r(val_X_test)
        test_loss = mse_loss(val_y_test, H_test)
    
    real_train_loss_history.append(train_loss.item())
    val_test_loss_history.append(test_loss.item())

    # if epoch % 10000 == 0:
    #     print(f'{epoch}/{epochs}: ', end='')
    #     W = l_r.weight.data.squeeze()
    #     b = l_r.bias.data
    #     for i, w in enumerate(W):
    #         print(f'W{i+1}: {w:.4f}, ', end='')
    #     print(f'b: {b.item():.4f}, Loss: {train_loss.item():.4f}\n')
    if epoch % 10000 == 0:
        print(f'{epoch}/{epochs}: ', end='')
        W = l_r.weight.data.squeeze()
        b = l_r.bias.data
        for i, w in enumerate(W):
            print(f'W{i+1}: {w:.4f}, ', end='')
        print(f'b: {b.item():.4f}, Loss: {test_loss.item():.4f}\n')
```

```python
plt.figure(figsize=(10, 5))
plt.plot(real_train_loss_history, label='Training Loss', linewidth=10, color= 'skyblue')
plt.plot(val_test_loss_history, label='Val Loss', linestyle='--' , linewidth=5, alpha=0.3, color='red')
plt.xlabel('Epoch')
plt.ylabel('MSE Loss')
plt.title('Loss Trend Over Epochs')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import torch
from torch.optim import SGD
from torch.nn.functional import mse_loss
from torch.nn import Linear

torch.manual_seed(321)


features, targets = pre_l_df.iloc[:,:-1], pre_l_df.iloc[:,-1]

X_train, X_test, y_train, y_test = \
train_test_split(features,targets, test_size=0.2, random_state=321)

real_X_train = torch.FloatTensor(X_train.values)
real_y_train = torch.FloatTensor(y_train.values).view(-1, 1)

val_X_test = torch.FloatTensor(X_test.values)
val_y_test = torch.FloatTensor(y_test.values).view(-1, 1)
l_r = Linear(real_X_train.shape[1], 1)  # ÏûÖÎ†• Ï∞®Ïõê ÎèôÏ†Å Ìï†Îãπ
optimizer = SGD(l_r.parameters(), lr=1e-3)  # Î≥¥Îã§ Ïã§Ïö©Ï†ÅÏù∏ ÌïôÏäµÎ•†

epochs = 100000  # Ï†ÅÏ†àÌïú ÏóêÌè¨ÌÅ¨ Ïàò
real_train_loss_history = []
val_test_loss_history = []

for epoch in range(1, epochs + 1):
    l_r.train()
    H = l_r(real_X_train)
    train_loss = mse_loss(real_y_train, H)

    optimizer.zero_grad()
    train_loss.backward()
    optimizer.step()
    
    l_r.eval()
    with torch.no_grad():
        H_test = l_r(val_X_test)
        test_loss = mse_loss(val_y_test, H_test)
        r2 = r2_score(val_y_test, H_test)
    
    real_train_loss_history.append(train_loss.item())
    val_test_loss_history.append(test_loss.item())

    # if epoch % 10000 == 0:
    #     print(f'{epoch}/{epochs}: ', end='')
    #     W = l_r.weight.data.squeeze()
    #     b = l_r.bias.data
    #     for i, w in enumerate(W):
    #         print(f'W{i+1}: {w:.4f}, ', end='')
    #     print(f'b: {b.item():.4f}, Loss: {train_loss.item():.4f}\n')
    if epoch % 1000 == 0:
        print(f'{epoch}/{epochs}: ', end='')
        W = l_r.weight.data.squeeze()
        b = l_r.bias.data
        for i, w in enumerate(W):
            print(f'W{i+1}: {w:.4f}, ', end='')
        print(f'b: {b.item():.4f}, Loss: {test_loss.item():.4f}\n')
        print(f'Test Loss: {test_loss.item():.4f}, R-squared: {r2.item():.4f}')
```

```python
plt.figure(figsize=(10, 5))
plt.plot(real_train_loss_history, label='Training Loss', linewidth=10, color= 'skyblue')
plt.plot(val_test_loss_history, label='Val Loss', linestyle='--' , linewidth=5, alpha=0.3, color='red')
plt.xlabel('Epoch')
plt.ylabel('MSE Loss')
plt.title('Loss Trend Over Epochs')
plt.legend()
plt.grid(True)
plt.show()
```
