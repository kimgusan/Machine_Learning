```python
# 데이터 가져오기
import pandas as pd

path = '/kdt_0900_kgs/ai/machine_learning/b_classifier/datasets/project_classifier/EEG_Eye_State_Classification.csv'
e_df = pd.read_csv(path)
e_df
```

```python
# 데이터 전처리 작업
# 1. 측정 위치에 따른 데이터 컬럼 위치 변경 후 copy
# 2. 결측치, 중복값 확인 
```

```python
# 원하는 컬럼 위치 변경 후 copy
re_columns = ['AF3', 'AF4', 'F3', 'F4', 'F7', 'F8', 'FC5', 'FC6',
    'T7', 'T8', 'P7', 'P8', 'O1', 'O2', 'eyeDetection']
e_need_df = e_df[re_columns].copy()
```

```python
# 결측치 확인
e_need_df.isna().sum()
```

```python
# 이상치 확인을 위한 시각화 그래프 확인
e_need_df.hist(figsize= (15,15))

# 특정 위치에 위치해있지만 minmax 값이 이상한 것으로 확인
# 따라서 이상치 제거 진행 후 히스토그램 한번 더 확인
```

```python
# Standardscaler 를 통한 이상치 제거
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
std_e_need = std.fit_transform(e_need_df.iloc[:, :-1])
```

```python
# 표준화를 시킨 데이터 프레임 생성 후 이상치 제거
std_e_need_df = pd.DataFrame(std_e_need, columns = e_need_df.iloc[:, :-1].columns)
std_e_need_df['eyeDetection'] = e_need_df.iloc[:, -1]
std_e_need_df.shape
```

```python
columns = std_e_need_df.iloc[:, :-1].columns
```

```python
# 이상치 제거
for column in columns:
    std_e_need_df.drop(
        std_e_need_df[~std_e_need_df[column].between(-1.96, 1.96)].index, 
        axis=0, 
        inplace=True)
    
for column in columns:
    print(f'{column}: {std_e_need_df[~std_e_need_df[column].between(-1.96, 1.96)].shape[0]}건')

std_e_need_df = std_e_need_df.reset_index(drop=True)
```

```python
# 이상치 제거 후 시각화 그래프 확인
std_e_need_df.hist(figsize= (15,15))
```

# 1. Cycle01. 결정트리 사용
- default: 이상치, 결측치, 중복값, 표준화, oversampling 작업 후 데이터 훈련 진행
  

```python
# 1. 데이터 훈련 (Decision Tree), SMOTE(Over-sampling 진행)
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from imblearn.over_sampling import SMOTE

features, targets = std_e_need_df.iloc[:, :-1], std_e_need_df.iloc[:, -1]

X_train, X_test, y_train, y_test = \
train_test_split(features, targets, stratify=targets, test_size=0.2, random_state=321)

# SMOTE 를 사용하여 Over Sampleing 진행
smote = SMOTE(random_state=321)
X_train_over, y_train_over = smote.fit_resample(X_train, y_train)

print(f'SMOTE 적용 전:\n{y_train.value_counts()}\n')
print(f'SMOTE 적용 후:\n{y_train_over.value_counts()}')
```

```python
std_e_need_df['eyeDetection'].unique()
```

```python
# 결정트리 데이터 훈련 진행
dtc = DecisionTreeClassifier()
dtc.fit(X_train_over, y_train_over)
```

```python
# 결정트리 이미지 추출을 위해 target 데이터를 문자열로 변경
import numpy as np

std_e_need_df['eyeDetection'] = std_e_need_df['eyeDetection'].astype(np.str_)
```

```python
# 결정트리 이미지 확인 후 dtc 의 최초 파라미터 확인
from sklearn.metrics import accuracy_score
print(dtc.max_depth)
print(dtc.min_samples_split)
print(dtc.feature_importances_)

# 교차 검증 이전 정확도 확인
prediction = dtc.predict(X_test)
acc_score = accuracy_score(y_test, prediction)
print('='* 50)
print(f"\n최초 정확도: {round(acc_score * 100,2)} %")
```

```python
# 교차 검증 진행 GridSearchCV
from sklearn.model_selection import GridSearchCV

parameters = {
    'max_depth': [5, 7, 9 ,10],
    'min_samples_split': [10, 20 ,30] 
}

g_dtc = GridSearchCV(dtc, param_grid=parameters, cv=3, refit=True, return_train_score=True)
g_dtc.fit(X_train_over, y_train_over)
```

```python
# 교차 검증 이후 정확도 확인 
print(g_dtc.best_params_, g_dtc.best_score_, sep='\n')
dtc = g_dtc.best_estimator_
prediction = dtc.predict(X_test)

acc_score = accuracy_score(y_test, prediction)
print('='* 50)
print(f"\n교차검증 이후 정확도: {round(acc_score * 100,2)} %")
```

```python
result_df = pd.DataFrame(g_dtc.cv_results_)[['params', 'mean_test_score', 'rank_test_score']]
display(result_df)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, ConfusionMatrixDisplay

def get_evaluation(y_test, prediction, classifier=None, X_test=None):
    confusion = confusion_matrix(y_test, prediction)
    accuracy = accuracy_score(y_test , prediction)
    precision = precision_score(y_test , prediction)
    recall = recall_score(y_test , prediction)
    f1 = f1_score(y_test, prediction)
    auc = roc_auc_score(y_test, prediction)
    
    print('오차 행렬')
    print(confusion)
    print('정확도: {0:.4f}, 정밀도: {1:.4f}, 재현율: {2:.4f}, F1: {3:.4f}, ROC-AUC: {4:.4f}'.format(accuracy, precision, recall, f1, auc))
    print("#" * 80)
    
    if classifier is not None and  X_test is not None:
        fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,4))
        titles_options = [("Confusion matrix", None), ("Normalized confusion matrix", "true")]

        for (title, normalize), ax in zip(titles_options, axes.flatten()):
            disp = ConfusionMatrixDisplay.from_estimator(classifier, X_test, y_test, ax=ax, cmap=plt.cm.Blues, normalize=normalize)
            disp.ax_.set_title(title)
        plt.show()
```

```python
get_evaluation(y_test, prediction, dtc, X_test)
```

```python
# 서포트 벡터 머신 모델을 이용한 데이터 훈련
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.svm import SVC

# 훈련데이터의 경우 기존에 SMOTE 메소드를 사용한 데이터를 사용

parameters = {
    'C': [0.001, 0.1, 1, 10, 100, 1000],
    'gamma': [0.001, 0.1, 1, 10, 100, 1000],
    'kernel': ['linear', 'rgb']
}

svc = SVC(probability=True)

g_svc = GridSearchCV(svc, param_grid=parameters, cv=3, refit=True, return_train_score=True)
g_svc.fit(X_train_over, y_train_over)

# 표준화의 경우 데이터셋이 모두 표준화 되어 있기 때문에 별도의 파이프라인 구축 없이 진행, 
# 분포도 표준 정규분포 형태를 띄고 있기 떄문에 MinMaxScaler 미진행
```

```python
# 최적의 파라미터 확인
result_df = pd.DataFrame(g_svc.cv_results_)[['params', 'mean_test_score', 'rank_test_score']]
display(result_df)

# 확인 결과 kernel 의 경우 rgb의 경우 값을 도출 할 수 없기 때문에 해당 값은 제외하고
# cost의 값과 gamma의 값이 높아지는 경우 수치값이 높은 것을 확인 할 수 있습니다.
# 현재 확인된 부분에 대하여 best_estimate 를 사용하여 훈련된 데이터에서 최적값을 찾을 수 있으나
# 차후 훈련 시 특정 조건에 대해서 세부적으로 다시 파라미터 값을 조절 할 수 있습니다.
```

### Cycle04. 하이퍼파라미터 값이 낮을수록 높은 점수를 보이고 있으며 3개의 훈련클래스에 대하여 측정됐을 때 가장 높은 수치를 보여주고 있습니다. 하지만 해당 훈련에 대해서는 횟수가 너무 적은 부분에 대해서는 높은 수치를 보이고 있기 때문에 추가적인 검증이 필요할 것으로 확인됩니다.

    과적합의 가능성
K 값의 영향: KNN에서 k 값이 너무 낮으면 모델이 훈련 데이터의 노이즈에 민감해져 과적합(overfitting)이 발생할 가능성이 높아집니다.   
k 값이 낮을 때 특정 데이터 포인트에 대해 매우 적은 수의 가장 가까운 이웃만을 고려하기 때문에,  
이는 결국 모델이 훈련 데이터에 지나치게 최적화되어 일반화 성능이 떨어질 수 있음을 의미합니다.    
과적합의 징후: 모델이 특정 훈련 클래스에서 높은 점수를 보이는 것은 좋은 성능처럼 보일 수 있으나, 
언급된 바와 같이 훈련 횟수가 적은 클래스에서도 높은 수치를 보인다면, 이는 모델이 과적합되어 있을 수 있는 신호입니다.   
특히, 적은 데이터 포인트로 훈련된 클래스에서 비현실적으로 높은 성능을 보이는 경우, 모델이 특정 패턴이나 노이즈를 학습했을 가능성이 있습니다.    

    추가 검증의 필요성   
검증 방법: 과적합을 확인하고 일반화 성능을 검증하기 위해서는 교차 검증(cross-validation) 같은 기법을 사용하는 것이 좋습니다.   
교차 검증을 통해 다양한 하위 데이터 세트에서 모델을 훈련하고 검증함으로써, 모델의 안정성과 일반화 능력을 더 정확히 평가할 수 있습니다.    
  
데이터 분할: 훈련 데이터와 검증 데이터를 적절히 분할하고, 모델이 새로운 데이터에 대해 어떻게 작동하는지 평가해야 합니다.   
또한, 훈련 데이터 내에서 불균형이 있는 경우, 이를 조정할 수 있는 기법을 적용할 필요가 있습니다.    

# 5. Cycle05. ensemble 사용 시 결정트리, svm, knn 의 모델에 대하여 voting 기능을 통한 최적의 값 확인

- default: 이상치, 결측치, 중복값, 표준화, oversampling 작업 후 데이터 훈련 진행
