```python
# under sampling ÏßÑÌñâ 
old_list = []
target_ = {}

# ÌäπÏÑ± Îç∞Ïù¥ÌÑ∞Îßå ÏÉòÌîåÎßÅ
for column in std_w_df.columns:
    if column != 'Target':
        sampled_data = std_w_df[column].sample(307019, random_state=321)
        old_list.append(sampled_data.rename(column))  # ÏÉòÌîåÎßÅÎêú Îç∞Ïù¥ÌÑ∞Î•º Î¶¨Ïä§Ìä∏Ïóê Ï∂îÍ∞Ä
    else:
        # 'Target' Ïª¨ÎüºÏùÄ Í∑∏ÎåÄÎ°ú Ïú†ÏßÄ
        target = std_w_df['Target']

# Î¶¨Ïä§Ìä∏Ïóê Ï†ÄÏû•Îêú Î™®Îì† ÏÉòÌîåÎßÅÎêú ÌäπÏÑ± Îç∞Ïù¥ÌÑ∞Î•º ÌïòÎÇòÏùò Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏúºÎ°ú Ìï©ÏπòÍ∏∞
under_w_df = pd.concat(old_list, axis=1).reset_index(drop=True)

# ÏÉòÌîåÎßÅÎêú ÌäπÏÑ± Îç∞Ïù¥ÌÑ∞Ïóê 'Target' Ïª¨Îüº Ï∂îÍ∞Ä
under_w_df['Target'] = target.reset_index(drop=True)

under_w_df
```

# 1. Cycle01. Í≤∞Ï†ïÌä∏Î¶¨ ÏÇ¨Ïö©
- default: Ïù¥ÏÉÅÏπò, Í≤∞Ï∏°Ïπò, Ï§ëÎ≥µÍ∞í, ÌëúÏ§ÄÌôî, oversampling ÏûëÏóÖ ÌõÑ Îç∞Ïù¥ÌÑ∞ ÌõàÎ†® ÏßÑÌñâ

```python
# 1. Îç∞Ïù¥ÌÑ∞ ÌõàÎ†® (Decision Tree), SMOTE(Over-sampling ÏßÑÌñâ)
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from imblearn.over_sampling import SMOTE

features, targets = under_w_df.iloc[:, :-1], under_w_df.iloc[:, -1]
targets

X_train, X_test, y_train, y_test = \
train_test_split(features, targets, stratify=targets, test_size=0.2, random_state=321)

# oversampling ÏßÑÌñâ
# smote = SMOTE(random_state=321)
# X_train_over, y_train_over = smote.fit_resample(X_train, y_train)

# print(f'SMOTE Ï†ÅÏö© Ï†Ñ:\n{y_train.value_counts()}\n')
# print(f'SMOTE Ï†ÅÏö© ÌõÑ:\n{y_train_over.value_counts()}')
```

```python
under_w_df['Target'].unique()
```

```python
# Í≤∞Ï†ïÌä∏Î¶¨ Îç∞Ïù¥ÌÑ∞ ÌõàÎ†® ÏßÑÌñâ
dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)
```

```python
# Í≤∞Ï†ïÌä∏Î¶¨ Ïù¥ÎØ∏ÏßÄ ÌôïÏù∏ ÌõÑ dtc Ïùò ÏµúÏ¥à ÌååÎùºÎØ∏ÌÑ∞ ÌôïÏù∏
from sklearn.metrics import accuracy_score
print(dtc.max_depth)
print(dtc.min_samples_split)
print(dtc.feature_importances_)

# ÍµêÏ∞® Í≤ÄÏ¶ù Ïù¥Ï†Ñ Ï†ïÌôïÎèÑ ÌôïÏù∏
prediction = dtc.predict(X_test)
acc_score = accuracy_score(y_test, prediction)
print('='* 50)
print(f"\nÏµúÏ¥à Ï†ïÌôïÎèÑ: {round(acc_score * 100,2)} %")


# ÏµúÏ¥à Ï†ïÌôïÎèÑÍ∞Ä ÎÜíÏùÄ ÏàòÏπòÎ•º Î≥¥Ïó¨ Í∏∞Ï°¥Ïóê Ï∞æÏïòÎçò ÌååÎùºÎØ∏ÌÑ∞ Í∞íÎ≥¥Îã§ ÎÜíÏùÄ ÏàòÏπòÎ•º 
# ÌïòÏù¥ÌçºÌååÎùºÎØ∏ÌÑ∞Î°ú ÏÑ§Ï†ïÌïòÏó¨ Í∞í Î∂ÑÎ•ò ÏßÑÌñâ 
```

```python
# ÍµêÏ∞® Í≤ÄÏ¶ù ÏßÑÌñâ GridSearchCV
from sklearn.model_selection import GridSearchCV

parameters = {
    'max_depth': [3, 5, 7],
    'min_samples_split': [10, 20 ,30] 
}

g_dtc = GridSearchCV(dtc, param_grid=parameters, cv=3, refit=True, return_train_score=True)
g_dtc.fit(X_train, y_train)
```

```python
# ÍµêÏ∞® Í≤ÄÏ¶ù Ïù¥ÌõÑ Ï†ïÌôïÎèÑ ÌôïÏù∏ 
print(g_dtc.best_params_, g_dtc.best_score_, sep='\n')
dtc = g_dtc.best_estimator_
prediction = dtc.predict(X_test)

acc_score = accuracy_score(y_test, prediction)
print('='* 50)
print(f"\nÍµêÏ∞®Í≤ÄÏ¶ù Ïù¥ÌõÑ Ï†ïÌôïÎèÑ: {round(acc_score * 100,2)} %")
```

```python
result_df = pd.DataFrame(g_dtc.cv_results_)[['params', 'mean_test_score', 'rank_test_score']]
display(result_df)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, ConfusionMatrixDisplay

def get_evaluation(y_test, prediction, classifier=None, X_test=None):
    confusion = confusion_matrix(y_test, prediction)
    accuracy = accuracy_score(y_test , prediction)
    precision = precision_score(y_test , prediction)
    recall = recall_score(y_test , prediction)
    f1 = f1_score(y_test, prediction)
    auc = roc_auc_score(y_test, prediction)
    
    print('Ïò§Ï∞® ÌñâÎ†¨')
    print(confusion)
    print('Ï†ïÌôïÎèÑ: {0:.4f}, Ï†ïÎ∞ÄÎèÑ: {1:.4f}, Ïû¨ÌòÑÏú®: {2:.4f}, F1: {3:.4f}, ROC-AUC: {4:.4f}'.format(accuracy, precision, recall, f1, auc))
    print("#" * 80)
    
    if classifier is not None and  X_test is not None:
        fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,4))
        titles_options = [("Confusion matrix", None), ("Normalized confusion matrix", "true")]

        for (title, normalize), ax in zip(titles_options, axes.flatten()):
            disp = ConfusionMatrixDisplay.from_estimator(classifier, X_test, y_test, ax=ax, cmap=plt.cm.Blues, normalize=normalize)
            disp.ax_.set_title(title)
        plt.show()
```

```python
get_evaluation(y_test, prediction, dtc, X_test)
```
