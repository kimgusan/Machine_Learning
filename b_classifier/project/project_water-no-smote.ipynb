```python
# under sampling 진행 
old_list = []
target_ = {}

# 특성 데이터만 샘플링
for column in std_w_df.columns:
    if column != 'Target':
        sampled_data = std_w_df[column].sample(307019, random_state=321)
        old_list.append(sampled_data.rename(column))  # 샘플링된 데이터를 리스트에 추가
    else:
        # 'Target' 컬럼은 그대로 유지
        target = std_w_df['Target']

# 리스트에 저장된 모든 샘플링된 특성 데이터를 하나의 데이터프레임으로 합치기
under_w_df = pd.concat(old_list, axis=1).reset_index(drop=True)

# 샘플링된 특성 데이터에 'Target' 컬럼 추가
under_w_df['Target'] = target.reset_index(drop=True)

under_w_df
```

# 1. Cycle01. 결정트리 사용
- default: 이상치, 결측치, 중복값, 표준화, oversampling 작업 후 데이터 훈련 진행

```python
# 1. 데이터 훈련 (Decision Tree), SMOTE(Over-sampling 진행)
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from imblearn.over_sampling import SMOTE

features, targets = under_w_df.iloc[:, :-1], under_w_df.iloc[:, -1]
targets

X_train, X_test, y_train, y_test = \
train_test_split(features, targets, stratify=targets, test_size=0.2, random_state=321)

# oversampling 진행
# smote = SMOTE(random_state=321)
# X_train_over, y_train_over = smote.fit_resample(X_train, y_train)

# print(f'SMOTE 적용 전:\n{y_train.value_counts()}\n')
# print(f'SMOTE 적용 후:\n{y_train_over.value_counts()}')
```

```python
under_w_df['Target'].unique()
```

```python
# 결정트리 데이터 훈련 진행
dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)
```

```python
# 결정트리 이미지 확인 후 dtc 의 최초 파라미터 확인
from sklearn.metrics import accuracy_score
print(dtc.max_depth)
print(dtc.min_samples_split)
print(dtc.feature_importances_)

# 교차 검증 이전 정확도 확인
prediction = dtc.predict(X_test)
acc_score = accuracy_score(y_test, prediction)
print('='* 50)
print(f"\n최초 정확도: {round(acc_score * 100,2)} %")


# 최초 정확도가 높은 수치를 보여 기존에 찾았던 파라미터 값보다 높은 수치를 
# 하이퍼파라미터로 설정하여 값 분류 진행 
```

```python
# 교차 검증 진행 GridSearchCV
from sklearn.model_selection import GridSearchCV

parameters = {
    'max_depth': [3, 5, 7],
    'min_samples_split': [10, 20 ,30] 
}

g_dtc = GridSearchCV(dtc, param_grid=parameters, cv=3, refit=True, return_train_score=True)
g_dtc.fit(X_train, y_train)
```

```python
# 교차 검증 이후 정확도 확인 
print(g_dtc.best_params_, g_dtc.best_score_, sep='\n')
dtc = g_dtc.best_estimator_
prediction = dtc.predict(X_test)

acc_score = accuracy_score(y_test, prediction)
print('='* 50)
print(f"\n교차검증 이후 정확도: {round(acc_score * 100,2)} %")
```

```python
result_df = pd.DataFrame(g_dtc.cv_results_)[['params', 'mean_test_score', 'rank_test_score']]
display(result_df)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, ConfusionMatrixDisplay

def get_evaluation(y_test, prediction, classifier=None, X_test=None):
    confusion = confusion_matrix(y_test, prediction)
    accuracy = accuracy_score(y_test , prediction)
    precision = precision_score(y_test , prediction)
    recall = recall_score(y_test , prediction)
    f1 = f1_score(y_test, prediction)
    auc = roc_auc_score(y_test, prediction)
    
    print('오차 행렬')
    print(confusion)
    print('정확도: {0:.4f}, 정밀도: {1:.4f}, 재현율: {2:.4f}, F1: {3:.4f}, ROC-AUC: {4:.4f}'.format(accuracy, precision, recall, f1, auc))
    print("#" * 80)
    
    if classifier is not None and  X_test is not None:
        fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,4))
        titles_options = [("Confusion matrix", None), ("Normalized confusion matrix", "true")]

        for (title, normalize), ax in zip(titles_options, axes.flatten()):
            disp = ConfusionMatrixDisplay.from_estimator(classifier, X_test, y_test, ax=ax, cmap=plt.cm.Blues, normalize=normalize)
            disp.ax_.set_title(title)
        plt.show()
```

```python
get_evaluation(y_test, prediction, dtc, X_test)
```
